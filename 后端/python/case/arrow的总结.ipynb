````python
import arrow
from datetime import datetime
````

# 1 获取当前时间

````python
# utc 时间
arrow.utcnow()
````

````python
# 获取 本地时间 
arrow.now()
````

注意上面的区别， now 是本地时，是有时区的 `+08:00` 

# 2 类型转换
## 2-1 arrow -> timestamp

````python
arrow.utcnow().timestamp
````

## 2-2 arrow -> str  
使用 `.format()`

````python
t=arrow.now()
````

````python
t
````

````python
t.format()
````

````python
t.format("YYYY-MM-DD HH:mm")
````

format 的参数如下:  
1- Y year  
2- M month
3- D day  
注意以上如是个位数会自动补全(最好使用此种方式)  
4- H hour  
5- m min

## 2-3 str -> arrow  
使用 `.get()` 方法

````python
# str 转换为 arrow
arrow.get("2020-07-04 20:48","YYYY-MM-DD HH:mm")
````

````python
# 时间戳 转换为 arrow
arrow.get(1593866814)
````

## 2-4 datetime -> arrow

````python
now=datetime(2020,10,1,1,1)
now
````

````python
arrow.get(now)
````

也可以使用 `fromdatetime` 进行转换

````python
arrow.fromdatetime(now)
````

# 3 时间计算/推移

对于 arrow 的推移，利于对当前的时间向前推 xx day，向后推 yy hour 等  
使用 `.shift()`  
可以使用的参数有
```
   days
   weeks
   months
   years
```

````python
t
````

````python
t.shift(days=1)
````

````python
t
````

注意使用 `.shift()` 会返回一个推移之后的时间，而不会修改当前的对象

# 4 其他

创建时间范围 ，使用 `range`

````python
start=arrow.Arrow(2020,2,1)
````

````python
end=arrow.Arrow(2020,5,1)
````

````python
for r in arrow.Arrow.range('hour', start, end):
    print(repr(r))
````
